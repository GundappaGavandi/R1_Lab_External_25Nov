### Find out the mean, median and variance for the column - DailyRate 

```python
df_mean = df['DailyRate'].mean()
df_median = df['DailyRate'].quantile(0.50)
df_variance_pop = df['DailyRate'].var()
df_variance_sample = sum(((df['DailyRate'] - df_mean)**2)/(len(df['DailyRate'])-1))
print("Mean of DailyRate = ",df_mean,"\n","Median of DailyRate = ", df_median,"\n",
      "Pop_Variance of DailyRate = ",df_variance_pop,"\n","Pop_Variance of DailyRate = ",df_variance_sample)
```
